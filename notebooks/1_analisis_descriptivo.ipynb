# Histograma de variables numericas

### AveragePrice (Precio Promedio)

#### Analisis
El precio promedio de venta del aguacate es de $1.41 USD, con una distribuciÃ³n sesgada hacia la derecha. La mayorÃ­a de los precios se concentran en el rango de $1.00 a $1.50, mientras que precios superiores a $2.00 son poco comunes. Esta tendencia sugiere que los aguacates son, en general, accesibles para los consumidores, con algunas excepciones que podrÃ­an estar relacionadas a factores externos como la oferta y la demanda.

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=avocado, x='AveragePrice', bins=30, kde=True, color='mediumseagreen')
plt.axvline(avocado['AveragePrice'].mean(), color='red', linestyle='--', label=f'Media: ${avocado["AveragePrice"].mean():.2f}')
plt.title('DistribuciÃ³n del Precio Promedio del Aguacate', fontsize=14)
plt.xlabel('Precio promedio (USD)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.legend()
plt.show()
```

### Total Volume (Volumen total vendido)

#### Analisis
El volumen total de ventas de aguacates presenta una distribuciÃ³n fuertemente sesgada a la derecha, donde la mayorÃ­a de las ventas diarias se sitÃºan entre 1,000 y 100,000 unidades. La aplicaciÃ³n de escala logarÃ­tmica permite observar mejor la presencia de valores extremos, correspondientes a eventos de alta demanda o ventas masivas. ### Total Bags (Bolsas totales)

#### Analisis
El anÃ¡lisis muestra una distribuciÃ³n multimodal en Total Bags, indicando tres grupos principales de compradores:

* PequeÃ±os minoristas (~100 bolsas): Compran volÃºmenes pequeÃ±os.

### Distribuciones de las variables numericas

```python
# variables numÃ©ricas para el grid de histogramas
numeric_vars = ['AveragePrice', 'Total Volume', '4046', '4225', '4770', 
                'Total Bags', 'Small Bags', 'Large Bags', 'XLarge Bags']

# Crear grid de histogramas
fig, axes = plt.subplots(3, 3, figsize=(18, 15))
axes = axes.ravel() # Aplanar la matriz de ejes para facilitar el acceso

for i, col in enumerate(numeric_vars):
    sns.histplot(data=avocado, x=col, bins=30, kde=True, ax=axes[i], color='firebrick')
    axes[i].set_title(f'DistribuciÃ³n de {col}', fontsize=10)
    axes[i].set_xlabel('')
    
    # Ajustar escalas para variables con amplio rango
    if col in ['Total Volume', 'Total Bags', '4046', '4225']:
        axes[i].set_xscale('log')

plt.tight_layout()
plt.suptitle('Distribuciones de las Variables NumÃ©ricas', y=1.02, fontsize=16)
plt.show()
```


## 1. DetecciÃ³n de Outliers con Boxplots

### 1.1. Boxplots Individuales para Variables Clave

**AnÃ¡lisis de Outliers:**
- **AveragePrice**:
  - El rango de precios se mantiene dentro de un rango saludable entre **$0.5** y **$3.25**.
  - Se observan pocos outliers, lo que indica una distribuciÃ³n relativamente estable.
  
- **Total Volume**:
  - Se presentan **miles de outliers extremos**, especialmente con valores superiores a **100M+** unidades, lo que sugiere la presencia de registros anÃ³malos de gran magnitud.
  
- **4046/4225**:
  - Ambas variables muestran **distribuciones similares**, aunque con **outliers significativos** que superan las **500k unidades**.
  
- **Total Bags**:
  - El comportamiento observado es **similar al de los volÃºmenes totales**, con algunos valores atÃ­picos que reflejan grandes cantidades de bolsas.

```python
# Variables numÃ©ricas a analizar
outlier_vars = ['AveragePrice', 'Total Volume', '4046', '4225', 'Total Bags']

# Colores personalizados para cada variable
colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightseagreen', 'lightsalmon']

plt.figure(figsize=(15, 8))
for i, col in enumerate(outlier_vars, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=avocado[col], color=colors[i-1])  # Asignar un color especÃ­fico a cada variable
    plt.title(f'Boxplot de {col}')
    if col != 'AveragePrice':  # Aplicar escala logarÃ­tmica solo a las variables de volumen
        plt.yscale('log')
plt.tight_layout()
plt.show()
```

### 1.2. Boxplot de Precio por Tipo y AÃ±o

**Hallazgos Clave:**
- **Diferencia de precios:**
  - Los aguacates **orgÃ¡nicos** son consistentemente **$1** mÃ¡s caros que los convencionales.
  
- **Tendencia temporal:**
  - El aÃ±o **2017** muestra los precios mÃ¡s altos para ambos tipos de aguacates.
  - Se observa una **distribuciÃ³n mÃ¡s estable** en los aÃ±os mÃ¡s recientes (**2016-2018**), lo que podrÃ­a indicar una mayor consistencia en los precios.

Este anÃ¡lisis mediante **boxplots** de precio por tipo y aÃ±o revela patrones significativos en la variabilidad de los precios de aguacates orgÃ¡nicos y convencionales a lo largo del tiempo.


```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=avocado, x='year', y='AveragePrice', hue='type', palette='YlGnBu')
plt.title('DistribuciÃ³n de Precios por AÃ±o y Tipo', fontsize=14)
plt.xlabel('AÃ±o', fontsize=12)
plt.ylabel('Precio Promedio (USD)', fontsize=12)
plt.legend(title='Tipo')
plt.show()
```

### 1.3. Outliers GeogrÃ¡ficos

**Insights Regionales:**
- **Mayor volumen:**
  - **California** y el **Noreste** lideran en tÃ©rminos de **volÃºmenes mÃ¡s altos** de aguacates.
  
- **DistribuciÃ³n compacta:**
  - **Chicago** muestra una **menor variabilidad** en sus datos, con una distribuciÃ³n mÃ¡s concentrada.

- **Casos atÃ­picos:**
  - La regiÃ³n **Southeast** presenta **outliers extremos**, lo que podrÃ­a sugerir la presencia de **eventos especiales** o situaciones inusuales en ciertos periodos.

```python
# Top 10 regiones por volumen de ventas
top_regions = avocado.groupby('region')['Total Volume'].sum().nlargest(10).index

plt.figure(figsize=(14, 7))
sns.boxplot(
    data=avocado[avocado['region'].isin(top_regions)], 
    x='region', 
    y='Total Volume',
    palette='Spectral'
)
plt.yscale('log')
plt.title('(Top 10) DistribuciÃ³n de Volumen por RegiÃ³n ', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.show()
```

## 2. AnÃ¡lisis de Correlaciones con Heatmaps

### 2.1. Matriz de CorrelaciÃ³n Completa

**Correlaciones Destacadas:**
- **Total Bags â Small Bags**: **0.99** *(relaciÃ³n casi perfecta)*.
- **Total Volume â 4046**: **0.98** *(el tamaÃ±o Hass grande domina el volumen total)*.
- **AveragePrice vs otras variables**:
  - Todas las correlaciones son menores a **|0.3|**, lo que indica una **correlaciÃ³n lineal insignificante** con otras variables.

Este anÃ¡lisis de correlaciones destaca las relaciones mÃ¡s fuertes entre las variables clave, mientras que sugiere que el **precio promedio** no muestra una relaciÃ³n significativa con otras variables.


```python
# Seleccionar variables numÃ©ricas
numeric_columns = ['AveragePrice', 'Total Volume', '4046', '4225', '4770', 
               'Total Bags', 'Small Bags', 'Large Bags', 'XLarge Bags']

# Calcular correlaciones
corr_matrix = avocado[numeric_columns].corr()

# Heatmap detallado
plt.figure(figsize=(14, 10))
sns.heatmap(
    corr_matrix, 
    annot=True, 
    cmap='viridis', 
    center=0, 
    fmt='.2f',
    linewidths=0.5,
    annot_kws={'size': 10}
)
plt.title('Matriz de CorrelaciÃ³n entre Variables NumÃ©ricas', fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

### 2.2. CorrelaciÃ³n Precio vs Volumen (Scatter Plot)

**PatrÃ³n Detectado:**
- **RelaciÃ³n inversa:**  
  Se observa una **correlaciÃ³n negativa significativa** (r â -0.4) entre el volumen y el precio.
  
- **Clusters identificados:**
  1. **Alto volumen + bajo precio:**
     - Productos **convencionales**.
     - Volumen tÃ­pico: >50k unidades.
     - Rango de precio: **$0.8-$1.5**.
  2. **Bajo volumen + alto precio:**
     - Productos **orgÃ¡nicos**.
     - Volumen tÃ­pico: <25k unidades.
     - Rango de precio: **$1.5-$3.0**.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=avocado.sample(1000),  # Muestra aleatoria para mejor visualizaciÃ³n
    x='Total Volume', 
    y='AveragePrice', 
    hue='type',
    alpha=0.6,
    palette='spring'
)
plt.xscale('log')
plt.title('RelaciÃ³n entre Precio y Volumen (Log Scale)', fontsize=14)
plt.xlabel('Volumen Total (log)', fontsize=12)
plt.ylabel('Precio Promedio (USD)', fontsize=12)
plt.legend(title='Tipo')
plt.show()
```
