### 1. An√°lisis Descriptivo del Dataset

Importacion de las librerias necesarias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print(plt.style.available)


plt.style.use('seaborn-v0_8')


%matplotlib inline
```

Se carga el dataset


```python
# Cargar datos
avocado = pd.read_csv('../data/avocado.csv')
avocado.head()
```

```python
# Estructura del dataset
avocado.info()

# Estad√≠sticas descriptivas
avocado.describe(include='all').transpose()
```

```python
# Eliminar columna 'Unnamed: 0' porque no aporta valor
avocado.drop(columns=['Unnamed: 0'], inplace=True)

# Convertir columna Date a tipo datetime
avocado['Date'] = pd.to_datetime(avocado['Date'])
```

# Histograma de variables numericas

### AveragePrice (Precio Promedio)

#### Analisis
El precio promedio de venta del aguacate es de $1.41 USD, con una distribuci√≥n sesgada hacia la derecha. La mayor√≠a de los precios se concentran en el rango de $1.00 a $1.50, mientras que precios superiores a $2.00 son poco comunes. Esta tendencia sugiere que los aguacates son, en general, accesibles para los consumidores, con algunas excepciones que podr√≠an estar relacionadas a factores externos como la oferta y la demanda.

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=avocado, x='AveragePrice', bins=30, kde=True, color='mediumseagreen')
plt.axvline(avocado['AveragePrice'].mean(), color='red', linestyle='--', label=f'Media: ${avocado["AveragePrice"].mean():.2f}')
plt.title('Distribuci√≥n del Precio Promedio del Aguacate', fontsize=14)
plt.xlabel('Precio promedio (USD)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.legend()
plt.show()
```

### Total Volume (Volumen total vendido)

#### Analisis
El volumen total de ventas de aguacates presenta una distribuci√≥n fuertemente sesgada a la derecha, donde la mayor√≠a de las ventas diarias se sit√∫an entre 1,000 y 100,000 unidades. La aplicaci√≥n de escala logar√≠tmica permite observar mejor la presencia de valores extremos, correspondientes a eventos de alta demanda o ventas masivas. Este patr√≥n sugiere que, aunque en general los vol√∫menes son moderados, existen situaciones extraordinarias que impulsan picos significativos en las ventas

```python
# Histograma de Total Volume en escala logar√≠tmica
plt.figure(figsize=(10,6))
sns.histplot(data=avocado, x='Total Volume', bins=50, kde=True, color='teal', log_scale=True)

# Formato del gr√°fico
plt.title('Distribuci√≥n del Volumen Total de Aguacates (Escala Logar√≠tmica)', fontsize=16)
plt.xlabel('Volumen Total (Escala Log)', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)
plt.grid(True)
plt.show()
```

### Total Bags (Bolsas totales)

#### Analisis
El an√°lisis muestra una distribuci√≥n multimodal en Total Bags, indicando tres grupos principales de compradores:

* Peque√±os minoristas (~100 bolsas): Compran vol√∫menes peque√±os.
* Distribuidores medianos (~10,000 bolsas): Adquieren vol√∫menes intermedios.
* Grandes cadenas (~1,000,000 bolsas): Compran grandes vol√∫menes.

Este patr√≥n refleja diferentes escalas de operaci√≥n en el mercado de aguacates, desde peque√±os minoristas hasta grandes cadenas de distribuci√≥n.

```python
plt.figure(figsize=(10, 6))
log_bins = np.logspace(0, 7, 50)  # Bins logar√≠tmicos
sns.histplot(data=avocado, x='Total Bags', bins=log_bins, kde=True, color='slateblue')
plt.xscale('log')
plt.title('Distribuci√≥n de Bolsas Totales (Escala Logar√≠tmica)', fontsize=14)
plt.xlabel('Total de bolsas (log scale)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.show()
```

### Distribuciones de las variables numericas

```python
# variables num√©ricas para el grid de histogramas
numeric_vars = ['AveragePrice', 'Total Volume', '4046', '4225', '4770', 
                'Total Bags', 'Small Bags', 'Large Bags', 'XLarge Bags']

# Crear grid de histogramas
fig, axes = plt.subplots(3, 3, figsize=(18, 15))
axes = axes.ravel() # Aplanar la matriz de ejes para facilitar el acceso

for i, col in enumerate(numeric_vars):
    sns.histplot(data=avocado, x=col, bins=30, kde=True, ax=axes[i], color='firebrick')
    axes[i].set_title(f'Distribuci√≥n de {col}', fontsize=10)
    axes[i].set_xlabel('')
    
    # Ajustar escalas para variables con amplio rango
    if col in ['Total Volume', 'Total Bags', '4046', '4225']:
        axes[i].set_xscale('log')

plt.tight_layout()
plt.suptitle('Distribuciones de las Variables Num√©ricas', y=1.02, fontsize=16)
plt.show()
```


## 1. Detecci√≥n de Outliers con Boxplots

### 1.1. Boxplots Individuales para Variables Clave

**An√°lisis de Outliers:**
- **AveragePrice**:
  - El rango de precios se mantiene dentro de un rango saludable entre **$0.5** y **$3.25**.
  - Se observan pocos outliers, lo que indica una distribuci√≥n relativamente estable.
  
- **Total Volume**:
  - Se presentan **miles de outliers extremos**, especialmente con valores superiores a **100M+** unidades, lo que sugiere la presencia de registros an√≥malos de gran magnitud.
  
- **4046/4225**:
  - Ambas variables muestran **distribuciones similares**, aunque con **outliers significativos** que superan las **500k unidades**.
  
- **Total Bags**:
  - El comportamiento observado es **similar al de los vol√∫menes totales**, con algunos valores at√≠picos que reflejan grandes cantidades de bolsas.

```python
# Variables num√©ricas a analizar
outlier_vars = ['AveragePrice', 'Total Volume', '4046', '4225', 'Total Bags']

# Colores personalizados para cada variable
colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightseagreen', 'lightsalmon']

plt.figure(figsize=(15, 8))
for i, col in enumerate(outlier_vars, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=avocado[col], color=colors[i-1])  # Asignar un color espec√≠fico a cada variable
    plt.title(f'Boxplot de {col}')
    if col != 'AveragePrice':  # Aplicar escala logar√≠tmica solo a las variables de volumen
        plt.yscale('log')
plt.tight_layout()
plt.show()
```

### 1.2. Boxplot de Precio por Tipo y A√±o

**Hallazgos Clave:**
- **Diferencia de precios:**
  - Los aguacates **org√°nicos** son consistentemente **$1** m√°s caros que los convencionales.
  
- **Tendencia temporal:**
  - El a√±o **2017** muestra los precios m√°s altos para ambos tipos de aguacates.
  - Se observa una **distribuci√≥n m√°s estable** en los a√±os m√°s recientes (**2016-2018**), lo que podr√≠a indicar una mayor consistencia en los precios.

Este an√°lisis mediante **boxplots** de precio por tipo y a√±o revela patrones significativos en la variabilidad de los precios de aguacates org√°nicos y convencionales a lo largo del tiempo.


```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=avocado, x='year', y='AveragePrice', hue='type', palette='YlGnBu')
plt.title('Distribuci√≥n de Precios por A√±o y Tipo', fontsize=14)
plt.xlabel('A√±o', fontsize=12)
plt.ylabel('Precio Promedio (USD)', fontsize=12)
plt.legend(title='Tipo')
plt.show()
```

### 1.3. Outliers Geogr√°ficos

**Insights Regionales:**
- **Mayor volumen:**
  - **California** y el **Noreste** lideran en t√©rminos de **vol√∫menes m√°s altos** de aguacates.
  
- **Distribuci√≥n compacta:**
  - **Chicago** muestra una **menor variabilidad** en sus datos, con una distribuci√≥n m√°s concentrada.

- **Casos at√≠picos:**
  - La regi√≥n **Southeast** presenta **outliers extremos**, lo que podr√≠a sugerir la presencia de **eventos especiales** o situaciones inusuales en ciertos periodos.

```python
# Top 10 regiones por volumen de ventas
top_regions = avocado.groupby('region')['Total Volume'].sum().nlargest(10).index

plt.figure(figsize=(14, 7))
sns.boxplot(
    data=avocado[avocado['region'].isin(top_regions)], 
    x='region', 
    y='Total Volume',
    palette='Spectral'
)
plt.yscale('log')
plt.title('(Top 10) Distribuci√≥n de Volumen por Regi√≥n ', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.show()
```

## 2. An√°lisis de Correlaciones con Heatmaps

### 2.1. Matriz de Correlaci√≥n Completa

**Correlaciones Destacadas:**
- **Total Bags ‚Üî Small Bags**: **0.99** *(relaci√≥n casi perfecta)*.
- **Total Volume ‚Üî 4225**: **0.85** *(el tama√±o Hass grande domina el volumen total)*.
- **AveragePrice vs otras variables**:
  - Todas las correlaciones son menores a **|0.3|**, lo que indica una **correlaci√≥n lineal insignificante** con otras variables.

Este an√°lisis de correlaciones destaca las relaciones m√°s fuertes entre las variables clave, mientras que sugiere que el **precio promedio** no muestra una relaci√≥n significativa con otras variables.


```python
# Seleccionar variables num√©ricas
numeric_columns = ['AveragePrice', 'Total Volume', '4046', '4225', '4770', 
               'Total Bags', 'Small Bags', 'Large Bags', 'XLarge Bags']

# Calcular correlaciones
corr_matrix = avocado[numeric_columns].corr()

# Heatmap detallado
plt.figure(figsize=(14, 10))
sns.heatmap(
    corr_matrix, 
    annot=True, 
    cmap='viridis', 
    center=0, 
    fmt='.2f',
    linewidths=0.5,
    annot_kws={'size': 10}
)
plt.title('Matriz de Correlaci√≥n entre Variables Num√©ricas', fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

### 2.2. Correlaci√≥n Precio vs Volumen (Scatter Plot)

**Patr√≥n Detectado:**
- **Relaci√≥n inversa:**  
  Se observa una **correlaci√≥n negativa significativa** (r ‚âà -0.4) entre el volumen y el precio.
  
- **Clusters identificados:**
  1. **Alto volumen + bajo precio:**
     - Productos **convencionales**.
     - Volumen t√≠pico: >50k unidades.
     - Rango de precio: **$0.8-$1.5**.
  2. **Bajo volumen + alto precio:**
     - Productos **org√°nicos**.
     - Volumen t√≠pico: <25k unidades.
     - Rango de precio: **$1.5-$3.0**.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=avocado.sample(1000),  # Muestra aleatoria para mejor visualizaci√≥n
    x='Total Volume', 
    y='AveragePrice', 
    hue='type',
    alpha=0.6,
    palette='spring'
)
plt.xscale('log')
plt.title('Relaci√≥n entre Precio y Volumen (Log Scale)', fontsize=14)
plt.xlabel('Volumen Total (log)', fontsize=12)
plt.ylabel('Precio Promedio (USD)', fontsize=12)
plt.legend(title='Tipo')
plt.show()
```
