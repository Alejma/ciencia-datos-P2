# Histograma de variables numericas

### AveragePrice (Precio Promedio)

#### Analisis
El precio promedio de venta del aguacate es de $1.41 USD, con una distribución sesgada hacia la derecha. La mayoría de los precios se concentran en el rango de $1.00 a $1.50, mientras que precios superiores a $2.00 son poco comunes. Esta tendencia sugiere que los aguacates son, en general, accesibles para los consumidores, con algunas excepciones que podrían estar relacionadas a factores externos como la oferta y la demanda.

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=avocado, x='AveragePrice', bins=30, kde=True, color='mediumseagreen')
plt.axvline(avocado['AveragePrice'].mean(), color='red', linestyle='--', label=f'Media: ${avocado["AveragePrice"].mean():.2f}')
plt.title('Distribución del Precio Promedio del Aguacate', fontsize=14)
plt.xlabel('Precio promedio (USD)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.legend()
plt.show()
```

### Total Volume (Volumen total vendido)

#### Analisis
El volumen total de ventas de aguacates presenta una distribución fuertemente sesgada a la derecha, donde la mayoría de las ventas diarias se sitúan entre 1,000 y 100,000 unidades. La aplicación de escala logarítmica permite observar mejor la presencia de valores extremos, correspondientes a eventos de alta demanda o ventas masivas. Este patrón sugiere que, aunque en general los volúmenes son moderados, existen situaciones extraordinarias que impulsan picos significativos en las ventas

```python
# Histograma de Total Volume en escala logarítmica
plt.figure(figsize=(10,6))
sns.histplot(data=avocado, x='Total Volume', bins=50, kde=True, color='teal', log_scale=True)

# Formato del gráfico
plt.title('Distribución del Volumen Total de Aguacates (Escala Logarítmica)', fontsize=16)
plt.xlabel('Volumen Total (Escala Log)', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)
plt.grid(True)
plt.show()
```

Se observa un pico cerca de 0, indicando una gran cantidad de registros con volúmenes muy bajos. Además, la distribución sigue una ley de potencias, característica de datos de ventas, donde una pequeña fracción de registros representa grandes volúmenes y la mayoría tiene volúmenes pequeños. Este patrón sugiere una concentración de ventas en volúmenes bajos, con algunos casos de alta demanda.

### 4225 - Total number of avocados with PLU 4225 sold (Hass grande)

### Total Bags (Bolsas totales)

#### Analisis
El análisis muestra una distribución multimodal en Total Bags, indicando tres grupos principales de compradores:

* Pequeños minoristas (~100 bolsas): Compran volúmenes pequeños.
* Distribuidores medianos (~10,000 bolsas): Adquieren volúmenes intermedios.
* Grandes cadenas (~1,000,000 bolsas): Compran grandes volúmenes.

Este patrón refleja diferentes escalas de operación en el mercado de aguacates, desde pequeños minoristas hasta grandes cadenas de distribución.

```python
plt.figure(figsize=(10, 6))
log_bins = np.logspace(0, 7, 50)  # Bins logarítmicos
sns.histplot(data=avocado, x='Total Bags', bins=log_bins, kde=True, color='slateblue')
plt.xscale('log')
plt.title('Distribución de Bolsas Totales (Escala Logarítmica)', fontsize=14)
plt.xlabel('Total de bolsas (log scale)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.show()
```

### Distribuciones de las variables numericas


## 1. Detección de Outliers con Boxplots

### 1.1. Boxplots Individuales para Variables Clave

**Análisis de Outliers:**
- **AveragePrice**:
  - El rango de precios se mantiene dentro de un rango saludable entre **$0.5** y **$3.25**.
  - Se observan pocos outliers, lo que indica una distribución relativamente estable.
  
- **Total Volume**:
  - Se presentan **miles de outliers extremos**, especialmente con valores superiores a **100M+** unidades, lo que sugiere la presencia de registros anómalos de gran magnitud.
  
- **4046/4225**:
  - Ambas variables muestran **distribuciones similares**, aunque con **outliers significativos** que superan las **500k unidades**.
  
- **Total Bags**:
  - El comportamiento observado es **similar al de los volúmenes totales**, con algunos valores atípicos que reflejan grandes cantidades de bolsas.

```python
# Variables numéricas a analizar
outlier_vars = ['AveragePrice', 'Total Volume', '4046', '4225', 'Total Bags']

# Colores personalizados para cada variable
colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightseagreen', 'lightsalmon']

plt.figure(figsize=(15, 8))
for i, col in enumerate(outlier_vars, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=avocado[col], color=colors[i-1])  # Asignar un color específico a cada variable
    plt.title(f'Boxplot de {col}')
    if col != 'AveragePrice':  # Aplicar escala logarítmica solo a las variables de volumen
        plt.yscale('log')
plt.tight_layout()
plt.show()
```

### 1.2. Boxplot de Precio por Tipo y Año

**Hallazgos Clave:**
- **Diferencia de precios:**
  - Los aguacates **orgánicos** son consistentemente **$1** más caros que los convencionales.
  
- **Tendencia temporal:**
  - El año **2017** muestra los precios más altos para ambos tipos de aguacates.
  - Se observa una **distribución más estable** en los años más recientes (**2016-2018**), lo que podría indicar una mayor consistencia en los precios.

Este análisis mediante **boxplots** de precio por tipo y año revela patrones significativos en la variabilidad de los precios de aguacates orgánicos y convencionales a lo largo del tiempo.


```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=avocado, x='year', y='AveragePrice', hue='type', palette='YlGnBu')
plt.title('Distribución de Precios por Año y Tipo', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Precio Promedio (USD)', fontsize=12)
plt.legend(title='Tipo')
plt.show()
```

### 1.3. Outliers Geográficos

**Insights Regionales:**
- **Mayor volumen:**
  - **California** y el **Noreste** lideran en términos de **volúmenes más altos** de aguacates.
  
- **Distribución compacta:**
  - **Chicago** muestra una **menor variabilidad** en sus datos, con una distribución más concentrada.

- **Casos atípicos:**
  - La región **Southeast** presenta **outliers extremos**, lo que podría sugerir la presencia de **eventos especiales** o situaciones inusuales en ciertos periodos.

```python
# Top 10 regiones por volumen de ventas
top_regions = avocado.groupby('region')['Total Volume'].sum().nlargest(10).index

plt.figure(figsize=(14, 7))
sns.boxplot(
    data=avocado[avocado['region'].isin(top_regions)], 
    x='region', 
    y='Total Volume',
    palette='Spectral'
)
plt.yscale('log')
plt.title('(Top 10) Distribución de Volumen por Región ', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.show()
```

## 2. Análisis de Correlaciones con Heatmaps

### 2.1. Matriz de Correlación Completa

**Correlaciones Destacadas:**
- **Total Bags ↔ Small Bags**: **0.99** *(relación casi perfecta)*.
- **Total Volume ↔ 4046**: **0.98** *(el tamaño Hass grande domina el volumen total)*.
- **AveragePrice vs otras variables**:
  - Todas las correlaciones son menores a **|0.3|**, lo que indica una **correlación lineal insignificante** con otras variables.

Este análisis de correlaciones destaca las relaciones más fuertes entre las variables clave, mientras que sugiere que el **precio promedio** no muestra una relación significativa con otras variables.


```python
# Seleccionar variables numéricas
numeric_columns = ['AveragePrice', 'Total Volume', '4046', '4225', '4770', 
               'Total Bags', 'Small Bags', 'Large Bags', 'XLarge Bags']

# Calcular correlaciones
corr_matrix = avocado[numeric_columns].corr()

# Heatmap detallado
plt.figure(figsize=(14, 10))
sns.heatmap(
    corr_matrix, 
    annot=True, 
    cmap='viridis', 
    center=0, 
    fmt='.2f',
    linewidths=0.5,
    annot_kws={'size': 10}
)
plt.title('Matriz de Correlación entre Variables Numéricas', fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

### 2.2. Correlación Precio vs Volumen (Scatter Plot)

**Patrón Detectado:**
- **Relación inversa:**  
  Se observa una **correlación negativa significativa** (r ≈ -0.4) entre el volumen y el precio.
  
- **Clusters identificados:**
  1. **Alto volumen + bajo precio:**
     - Productos **convencionales**.
     - Volumen típico: >50k unidades.
     - Rango de precio: **$0.8-$1.5**.
  2. **Bajo volumen + alto precio:**
     - Productos **orgánicos**.
     - Volumen típico: <25k unidades.
     - Rango de precio: **$1.5-$3.0**.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=avocado.sample(1000),  # Muestra aleatoria para mejor visualización
    x='Total Volume', 
    y='AveragePrice', 
    hue='type',
    alpha=0.6,
    palette='spring'
)
plt.xscale('log')
plt.title('Relación entre Precio y Volumen (Log Scale)', fontsize=14)
plt.xlabel('Volumen Total (log)', fontsize=12)
plt.ylabel('Precio Promedio (USD)', fontsize=12)
plt.legend(title='Tipo')
plt.show()
```
