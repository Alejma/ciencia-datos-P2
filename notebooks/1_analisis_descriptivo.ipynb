### 1. An√°lisis Descriptivo del Dataset

Importacion de las librerias necesarias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print(plt.style.available)


plt.style.use('seaborn-v0_8')


%matplotlib inline
```

Se carga el dataset


```python
# Cargar datos
avocado = pd.read_csv('../data/avocado.csv')
avocado.head()
```

Primer vistazo al data set (Informacion basica)

```python
# Estructura del dataset
avocado.info()

# Estad√≠sticas descriptivas
avocado.describe(include='all').transpose()
```

Limpieza inicial y ajustes

```python
# Eliminar columna 'Unnamed: 0' porque no aporta valor
avocado.drop(columns=['Unnamed: 0'], inplace=True)

# Convertir columna Date a tipo datetime
avocado['Date'] = pd.to_datetime(avocado['Date'])
```

# Histograma de variables numericas

### AveragePrice (Precio Promedio)

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=avocado, x='AveragePrice', bins=30, kde=True, color='mediumseagreen')
plt.axvline(avocado['AveragePrice'].mean(), color='red', linestyle='--', label=f'Media: ${avocado["AveragePrice"].mean():.2f}')
plt.title('Distribuci√≥n del Precio Promedio del Aguacate', fontsize=14)
plt.xlabel('Precio promedio (USD)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.legend()
plt.show()
```
